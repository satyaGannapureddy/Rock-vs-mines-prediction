Handlign imbalance datapoints in datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('credit_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#distribution of 2 classes
df['Class'].value_counts()
```

This is highly imbalance dataset,
0-> legit transaction
1-> fraud transaction

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#undersampling
#we are building a sample dataset containing similar distibution of fradulent
#transactions
```

```python
legit_sample= legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```
