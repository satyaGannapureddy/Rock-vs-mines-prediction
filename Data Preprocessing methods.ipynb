```python
print(data)
```

```python
df.head()
```

```python
df.shape
```

```python
x=df
y=data.target
```

```python
print(x)
```

```python
print(y)
```

Label Encoding:

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
cancer_data=pd.read_csv('breast cancer.csv')
```

```python
cancer_data
```

```python
cancer_data['diagnosis'].value_counts()
```

```python
label_encode=LabelEncoder()
```

```python
labels=label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to dataframe
cancer_data['target']=labels
```

```python
cancer_data.head()
```

```python
#0-benine
#1-malignant
cancer_data['target'].value_counts()
```
